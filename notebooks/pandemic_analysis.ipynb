## Problem Statement

Pandemics require early risk assessment to support healthcare systems and policy decisions.  
This project analyzes pandemic case trends using data visualization and predictive modeling techniques inspired by my published research work.

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading Dataset

```python
df = pd.read_csv("D:/github/pandemic-risk-analyzer/data/covid_19_clean_complete.csv")
df.head()
```

## Dataset Overview

```python
df.info()
df.describe()
```

## Data Cleaning

```python
df = df.dropna()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

## Global Pandemic Trend Analysis
Aggregating worldwide confirmed cases over time.

```python
global_cases = df.groupby("Date")["Confirmed"].sum()
global_cases.head()
```

## Trend Visualization

```python
global_cases.plot(figsize=(10,5))

plt.title("Global Pandemic Case Trend")
plt.xlabel("Date")
plt.ylabel("Confirmed Cases")

plt.savefig("../images/global_cases.png")

plt.show()
```

## Death Trend Analysis
Tracking global mortality progression.

```python
global_deaths = df.groupby("Date")["Deaths"].sum()

global_deaths.plot(figsize=(10,5))

plt.title("Global Death Trend")
plt.xlabel("Date")
plt.ylabel("Deaths")

plt.savefig("../images/death_trend.png")

plt.show()
```

## Recovery Ananlysis

```python
global_recovered = df.groupby("Date")["Recovered"].sum()

global_recovered.plot(figsize=(10,5))

plt.title("Global Recovery Trend")
plt.xlabel("Date")
plt.ylabel("Recovered")

plt.savefig("../images/recovery_trend.png")

plt.show()
```

## Pandemic Risk Index
A simple metric estimating outbreak severity.

```python
risk_index = (global_cases.diff()) / (global_recovered.diff() + 1)

risk_index.plot(figsize=(10,5))

plt.title("Pandemic Risk Index")
plt.xlabel("Date")
plt.ylabel("Risk Level")

plt.savefig("../images/risk_index.png")

plt.show()
```

## Conclusion

The analysis demonstrates how infection growth, mortality,
and recovery dynamics evolve during a pandemic.

Such analytics can help policymakers and healthcare systems
anticipate risk escalation.
